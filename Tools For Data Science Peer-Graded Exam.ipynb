### The following python code is the implementation of the kinetic energy formula

```python
# The formula for kinetic energy is given by 1/2 times mass times velocity to the power of 2
def kinetic_energy(m, v):
    k = 0.5 * m * v**2
    print("Kinetic Energy: ", k )

kinetic_energy(10, 2)
```
